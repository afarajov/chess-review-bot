```python
import requests

username = "konteyni"

def fetch_games(username):
    headers = {'User-Agent': 'my-analysis-bot/1.0 (contact: your-email@example.com)'}
    url = f"https://api.chess.com/pub/player/{username}/games/2026/06"

    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        games_data = response.json()
        games = games_data.get("games", [])
        return games
    else:
        print(f"Failed to fetch games for {username}. Status code: {response.status_code}")
        return []
    
fetched_games = fetch_games(username)

fetched_games
```
